```python
import os
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("groq_api_key")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.3-70b-versatile", api_key=api_key)
print(model)
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="Hello my name is Raviraj ! what's your name ")])
```

Message History 

we can use message history to  wrap the chat model and make it stateful this will keep track of input and outputs of the model and store them into some Datastore future intractions then load this mesages  and pass them into the chains as part of the input 

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}


def get_session_history(session_id : str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model,get_session_history)
```

```python
config = {"configurable":{"session_id" :"chat1"}}
```

```python
response =with_message_history.invoke(
    [HumanMessage(content="Hello my name is Raviraj ! ?")],
    config=config
)
```

```python
response =with_message_history.invoke(
    [HumanMessage(content="do you able to recall my name ?")],
    config=config
)
print(response.content)
```
